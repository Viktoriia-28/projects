```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[]
i=1
for i in range(53):
    i=i+1
    x.append(i)
```

```python
data=pd.DataFrame(data=x, columns=["Неделя"])
```

```python
def kolihestvo(row):
    nedel=row['Неделя']
    if nedel> 20 and nedel< 31:
        kol=7200
    else:
        kol=12000
    return kol
def price(row):
    nedel=row['Неделя']
    if nedel> 20:
        price1=11.5
    else:
        price1=12.68
    return price1
data['Цена за коробку']=data.apply(price, axis=1)
data['Количество']=data.apply(kolihestvo, axis=1)
data['Закупка']=data['Количество']*9
data['Фрахт']=data['Количество']/1200*3000
data['Продажа']=data['Количество']*data['Цена за коробку']
data['Прибыль']=data['Продажа']-data['Закупка']-data['Фрахт']
```

```python
data
```

```python
plt.figure(figsize=(15,5))
plt.plot(data['Неделя'], data['Продажа'], label='Выручка')
plt.plot(data['Неделя'], data['Закупка']+data['Фрахт'], label='Сумма затрат на закупку продукции и логистику')
plt.title('Обороты по продажам и закупке')
ax=plt.gca()
ax.set_xlabel('Неделя')
ax.set_ylabel('Оборот, $')
plt.title('Обороты затрат и выручки за год')
plt.legend() 
```

Таким образом , мы получаем прибыль в первые 20 недель.
